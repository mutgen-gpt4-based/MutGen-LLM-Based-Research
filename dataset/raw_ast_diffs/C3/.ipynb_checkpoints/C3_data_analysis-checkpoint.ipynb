```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
caminho_arquivo = "C3_buggy_AST.txt"
with open(caminho_arquivo, "r") as fhandle:
    dataset = fhandle.read()

complete_ast = []

dataset = dataset.split("LINE:")
for ast_unclean in dataset:
    if not ast_unclean:
        continue
    all_actions = []
    line = ast_unclean.split("###")[0].strip()
    actions = ast_unclean.split("###")[1].split(" ||| ") 
    print(line, actions)

    for ast_mod in actions:
        if "Example\n\tExample" in ast_mod:
            continue
        if "from Example:1" in ast_mod:
            all_actions.append(ast_mod.split(" from ")[0].strip())

        if "at Example:1" in ast_mod:
            all_actions.append(ast_mod.split(" at ")[0].strip())
    clean_ast = (line, all_actions)
    if "\n" in clean_ast[-1]:
        drop = clean_ast.pop(-1)
    complete_ast.append(clean_ast)
```

```python
def analisar_distribuicao_diffs(dados):
    all_nodes = []
    for tupla in dados:
        for mod_tp in tupla[1]:
            node = mod_tp.split(" ")[1]
            all_nodes.append(node)

    np_subjects_dataset = np.array(all_nodes)
    valores_unicos, contagens = np.unique(np_subjects_dataset, return_counts=True)
    valores_unicos, contagens = np.unique(np_subjects_dataset, return_counts=True)
    print("Array original:", np_subjects_dataset)
    print("Valores únicos:", valores_unicos)
    print("Contagens:", contagens)
    print("TOTAL DE NODES: ", len(valores_unicos))

    media_contagens = np.mean(contagens)
    desvio_padrao_contagens = np.std(contagens)

    print(f"\nMédia das Contagens: {media_contagens:.2f}")
    print(f"Desvio Padrão das Contagens: {desvio_padrao_contagens:.2f}")
    print("\n--- Frequência dos valores ---")
    for valor, contagem in zip(valores_unicos, contagens):
        print(f"Valor {valor}: {contagem} vezes")
    
    plt.figure(figsize=(10, 6)) 
    plt.bar(valores_unicos, contagens, color='skyblue') 

    plt.xlabel("Valores") 
    plt.ylabel("Contagem de Ocorrências") 
    plt.title("Frequência dos Valores no Array") 
    plt.xticks(valores_unicos) 
    plt.grid(axis='y', linestyle='--', alpha=0.7) 
    plt.tight_layout() 
    plt.show()

    ##################################################################################

    df1 = pd.DataFrame(complete_ast, columns=['id_linha', 'acoes']) 
    df1['num_acoes'] = df1['acoes'].str.len()
    distribuicao_tamanho = df1['num_acoes'].value_counts().sort_index()

    print("Tabela de Distribuição (Nº de Ações por Linha):")
    print(distribuicao_tamanho)
    print("-" * 30)

    plt.figure(figsize=(10, 6))
    sns.barplot(x=distribuicao_tamanho.index, y=distribuicao_tamanho.values, palette='mako')
    plt.title('Distribuição da Quantidade de Ações por Linha', fontsize=16)
    plt.xlabel('Número de Ações na Linha', fontsize=12)
    plt.ylabel('Contagem de Linhas (Frequência)', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    ####################################################################################

    estatisticas = df1['num_acoes'].describe()
    print(estatisticas)
    print("======================================================")
    uniques = df1.explode('acoes')["acoes"].unique()
    unbalance_dataset = zip(valores_unicos.tolist(), contagens.tolist())

    return uniques, unbalance_dataset
```

```python
uniques, unbalance_dataset = analisar_distribuicao_diffs(complete_ast)
```

### Cutting dataset

```python
unbalance_data = dict(unbalance_dataset)
```

```python
def filter_complete_ast_by_frequency(all_ops_and_freq, complete_ast_list, min_frequency=100):
    below_threashhold = []

    for operator in all_ops_and_freq.keys():
        if all_ops_and_freq[operator] < min_frequency:
            below_threashhold.append(operator)
            
    return below_threashhold

cutted_lines_threashold = filter_complete_ast_by_frequency(unbalance_data, complete_ast)
print("Numero de nodes cortados: ", len(cutted_lines_threashold))
```

### Criação do dataset balanceado

<hr>

```python
balance_dataset = []
for line, only_operators in complete_ast:
    condition = False
    print("linha ", line, 'operators ', only_operators)
    for operator in only_operators:
        if operator.split()[1].strip() in cutted_lines_threashold:
            condition = True
            print("achei")
    if condition == False:
        balance_dataset.append((line, only_operators))
```
